# Assessment 2
A school has a simple program to track student names.  It loads up a file with existing names and then allows the user to add new names.  These new names should then be saved to the file so that next time the program is run the entered students are available. 

The school wishes to run their application from Google CoLab, so the code must be compatible with that environment.

This is accessed via a simple console menu system.  The code is provided below.

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

def init():
    try:
        with open("/content/drive/My Drive/students.txt", 'r') as file:
            pass
    except FileNotFoundError:
        with open("/content/drive/My Drive/students.txt", 'w') as file:
            file.write("Sid Whardley\n")
            file.write("Barth Lawrenz\n")
            file.write("Chrissy Bradnum\n")
            file.write("Brigida Andriuzzi\n")
            file.write("Siegfried Lamplugh\n")
            file.write("Eunice Parlet\n")
            file.write("Bobbi Lamps\n")
            file.write("Clayborne Nutley\n")
            file.write("Rivalee Laight\n")
            file.write("Trish Adams\n")

# Function to load existing names from a file
def load_names():
    with open("/content/drive/My Drive/students.txt", 'r') as file:
        names = file.read()

    print(names)

# Function to add a student
def add_name():
    new_name = input("Enter the new student's name: ")


# Function to display the menu
def display_menu():
    print("\nStudent Tracker Menu:")
    print("1. View all names")
    print("2. Add a new name")
    print("3. Exit")

# Main program
def student_tracker():
    while True:
        display_menu()
        choice = input("Enter your choice: ")

        if choice == '1':
            print("\nStudent Names:")
            load_names()

        elif choice == '2':
            add_name()

        elif choice == '3':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

init()
# Run the program
student_tracker()
```
